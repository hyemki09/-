```python
import urllib.request
from bs4 import BeautifulSoup
```

```python
url='https://search.naver.com/search.naver?where=nexearch&sm=top_hty&fbm=1&ie=utf8&query=%EC%98%81%ED%99%94%EC%88%9C%EC%9C%84'
req = urllib.request.urlopen(url) # url에 대한 연결요청
res = req.read()
```

```python
res
```

```python
soup = BeautifulSoup(res,'html.parser')
keywords = soup.find_all('strong',class_='name')
```

```python
keywords
```

```python
keywords = [a.get_text().strip() for a in keywords[:10]]
```

```python
keywords
```

```python
customer = soup.find_all('span',class_='sub_text')
```

```python
customer
```

```python
customer=[a.get_text().strip() for a in customer[:10]]
```

```python
customer
```

```python
keywords.append(customer)
```

```python
for x,y in zip(keywords,customer):
    print(x,y)
```
